# Question 2. Is it possible to predict 'detail views' from the information given in the other columns?

As per my intial analysis and the description provided for the columns. 
I would have to say, it is definitely possible to predict 'detail views' with the help of other columns, particularly; column 'search_views' & 'ctr'.
Given the explaination for column ctr, "Click through rate calculated as the quotient of detail_views over search_views"
detail_views can be calculated (predicted) by the simple multiplication of ctr and search_views
detail_views = ctr * search_views

To compare my calculated result with the original, I will show a demonstration

```python
import pandas as pd
from sklearn import preprocessing

df = pd.read_csv('AS24_Case_Study_Data_Processed_2.csv')
Mapping_Tier = {"Basic":0, "Plus":1, "Premium":2}
df["product_tier"] = df["product_tier"].replace(Mapping_Tier)

from sklearn import preprocessing

def encode_labels(df):
    le = preprocessing.LabelEncoder()
    for column in df.columns:
        if df[column].dtype == type(object):
         df[column] = le.fit_transform(df[column])
    return df

df = encode_labels(df)

training_df = df.sample(frac=0.7,random_state=200) 
test_df = df.drop(training_df.index)
```

```python
detail_views = training_df["detail_views"]
training_df = training_df.drop('detail_views', axis=1)
test_df = test_df.drop('detail_views', axis=1)
training_df = training_df.drop(['created_date','deleted_date','article_id'], axis=1)
test_df = test_df.drop(['created_date','deleted_date','article_id'], axis=1)
```

Support Vector Regression

Multi-layer Perceptron Regressor

```python
from sklearn.neural_network import MLPRegressor

modelMLP = MLPRegressor(hidden_layer_sizes=(21), max_iter=1000)
modelMLP.fit(training_df, detail_views)

res = modelMLP.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelMLP.score(training_df, detail_views)
print("Score Multi-layer Perceptron Regressor: " + str(r_square))
```

Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(max_depth=16)
modelRFR = regressor.fit(training_df, detail_views)

res = modelRFR.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelRFR.score(training_df, detail_views)
print("Score Random Forest Regressor: " + str(r_square))
```

Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(max_depth=16)
modelDTR = regressor.fit(training_df, detail_views)

res = modelDTR.predict(test_df)
predicted_df = pd.Series(res)

from sklearn.metrics import r2_score
r_square = modelDTR.score(training_df, detail_views)
print("Score Decision Tree Regressor: " + str(r_square))
```

Linear Regression

For evaluation on such problems, a R-Squared score and a mean-squared score is very important.
Comparing models on the R-squared metric will tell us how closely accurate the predictions could be.
Higher R-squared values represent smaller differences between the observed data and the fitted values.

A Decision Tree Regressor model gave me the best R-squared score (99.99%) compared to any other model.
